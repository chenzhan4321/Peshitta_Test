```python
import re

def tokenize(text):
    return text.split()

tokens = tokenize(text)
```

```python

def mapping(tokens):
    word_to_id = {}
    id_to_word = {}
    
    for i, token in enumerate(set(tokens)):
        word_to_id[token] = i
        id_to_word[i] = token
    
    return word_to_id, id_to_word

word_to_id, id_to_word = mapping(tokens)
print (word_to_id)
```

```python
import numpy as np

np.random.seed(42)

def generate_training_data(tokens, word_to_id, window, chunk_size=10000):
    X = []
    y = []
    n_tokens = len(tokens)
    chunk_index = 0
    example_count = 0
    
    for i in range(n_tokens):
        print(i) if i % 1000 == 0 else None
        idx = concat(
            range(max(0, i - window), i), 
            range(i, min(n_tokens, i + window + 1))
        )
        for j in idx:
            if i == j:
                continue
            X.append(one_hot_encode(word_to_id[tokens[i]], len(word_to_id)))
            y.append(one_hot_encode(word_to_id[tokens[j]], len(word_to_id)))
            example_count += 1
            
            if example_count == chunk_size:
                save_chunk(X, y, chunk_index)
                X, y = [], []
                chunk_index += 1
                example_count = 0
    
    # Save the last chunk if it's not empty
    if example_count > 0:
        save_chunk(X, y, chunk_index)

def concat(*iterables):
    for iterable in iterables:
        yield from iterable
        
def one_hot_encode(id, vocab_size):
    res = [0] * vocab_size
    res[id] = 1
    return res

def save_chunk(X, y, chunk_index):
    X = np.asarray(X)
    y = np.asarray(y)
    np.save(f'/home/zhanchen/Documents/X_chunk_{chunk_index}.npy', X)
    np.save(f'/home/zhanchen/Documents/y_chunk_{chunk_index}.npy', y)
    print(f"Saved chunk {chunk_index}")

X, y = generate_training_data(tokens, word_to_id, 2)
```

```python
def init_network(vocab_size, n_embedding):
    model = {
        "w1": np.random.randn(vocab_size, n_embedding),
        "w2": np.random.randn(n_embedding, vocab_size)
    }
    return model

model = init_network(len(word_to_id), 100)
```

```python
import torch
import torch.nn.functional as F

#forward pass

def forward(model, X, return_cache=True):
    cache = {}
    
    cache["a1"] = X @ model["w1"]
    cache["a2"] = cache["a1"] @ model["w2"]
    cache["z"] = F.softmax(torch.tensor(cache["a2"], dtype=torch.float), dim=-1).numpy()
    
    if not return_cache:
        return cache["z"]
    return cache

def backward(model, X, y, alpha):
    cache  = forward(model, X)
    da2 = cache["z"] - y
    dw2 = cache["a1"].T @ da2
    da1 = da2 @ model["w2"].T
    dw1 = X.T @ da1
    assert(dw2.shape == model["w2"].shape)
    assert(dw1.shape == model["w1"].shape)
    model["w1"] -= alpha * dw1
    model["w2"] -= alpha * dw2
    return cross_entropy(cache["z"], y)

def cross_entropy(z, y):
    epsilon = 1e-15
    z = np.clip(z, epsilon, 1. - epsilon)
    return -np.sum(y * np.log(z))
```

```python


import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'svg'

n_iter = 100
learning_rate = 0.05

history = []

for count in range(n_iter):
    print(count)
    result = backward(model, X, y, learning_rate)
    history.append(result)

plt.plot(range(len(history)), history, color="skyblue")
plt.show()
```
